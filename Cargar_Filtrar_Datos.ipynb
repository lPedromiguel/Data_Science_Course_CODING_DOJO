```python
import pandas as pd
```

```python
bike_df = '/content/drive/MyDrive/Data science/Pandas/RailsToTrails_National Count Data_week.xlsx'
df = pd.read_excel(bike_df)
df.head()
```

```python
date_filter = df['Week of'] > '2020-07-31'
```

```python
Filtered_DF = df[date_filter]
Filtered_DF
```

```python
change_filter = df['Change 2019-2020'] >= float('1')
```

```python
Filtered_DF_2 = df[change_filter]
Filtered_DF_2
```

```python
Filtered_DF_Final = Filtered_DF[change_filter]
Filtered_DF_Final
```
